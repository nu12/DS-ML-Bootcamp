# K Means Clustering

K Means Clustering is an unsupervised learning algorithm that will attempt to group similar cluster together in your data. The overall goal is to divide data into distinct groups sucj that observations within each group are similar.

The algorithm works as the following:
* Choose a numer of clusters "k"
* Randomly assign each point to a cluster
* Until clusters stop changing:
    * for each cluster, compute the centroid by takin the mean vector distance of the cluster points.
    * Re-assign each data point to the cluster for which the centroid is the closest.
    
![](https://datascienceplus.com/wp-content/uploads/2019/09/Capture.png)

## Choosing a K value

The elbow method is a way to define it, but in general there is no easy answer for choosing the best "k" value. It also depends on domain expertise.

### Elbow method

Compute the sum of squared error (SSE)* for some values of k.

Plot k against the SSE.

The idea is to choose the latest k at which the SSE decreases abruptly.

![](https://www.oreilly.com/library/view/statistics-for-machine/9781788295758/assets/995b8b58-06f1-4884-a2a1-f3648428e947.png)

(*) SSE is defined as the sum of the squared distance between each member of the cluster and its centroid.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate artificial data
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, 
                  n_features=2, 
                  centers=4, 
                  cluster_std=1.8, 
                  random_state=101)
# data is a tuple, the first element are the features and the second is the cluster the observation belongs to

plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
# Center of each cluster
km.cluster_centers_
```

```python
# Labels calculated by the algorithm
# In a real world scenario, we would not have the actual cluster for each observation
# This would be the final output of the process
km.labels_
```

```python
# Because we know the clustering generated bu the artificial data
# we can compare the results

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c=km.labels_, cmap="rainbow")

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```
