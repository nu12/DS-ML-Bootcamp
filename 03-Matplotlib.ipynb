# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Show plots without plt.show()
%matplotlib inline 
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

## Functional method

```python
# Basic plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# Plot more than 1 figure
plt.subplot(1,2,1) # Row, columns, plot member
plt.plot(x,y, 'r-')

plt.subplot(1,2,2) # Row, columns, plot member
plt.plot(y,x, 'g-')
```

## Object oriented method
Create figure objects and call methods to manipulate it.

```python
# fig is like a blank canvas
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

### Multiple axes in one plot

```python
# Demontration of add_axes method
fig = plt.figure()

# add_axes([ left, bottom, width, height ])
axes1 = fig.add_axes([0.0,0.0,1,1])
axes2 = fig.add_axes([0.1,0.1,0.8,0.8])
axes3 = fig.add_axes([0.2,0.3,0.5,0.4])
```

```python
# Ploting two axes
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.0,1,1])
axes2 = fig.add_axes([0.1,0.6,0.3,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

### Subplots (multiple plots)

```python
# Using subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# axes is an array
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout() # Improves layout
```

#### Legend

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2, label="Squared")
axes.plot(x,x**3, label="Cubed")

axes.legend(loc=0) # Use location code to change location
```

```python
axes.legend(loc=(0.05,0.1)) # Use a tutple to specify the location
fig
```

#### Line color and style

```python
# Plot appearance
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Color name
ax.plot(x,y, color='green')
```

```python
# Color HEX code
ax.plot(x,y, color='#af0951')
fig
```

```python
# Line width & alpha
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#a4a832', lw=5, alpha=0.5) # lw = linewidth
```

```python
# Line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#a4a832', lw=1, ls="--") # ls= linestyle
# supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# Markers
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='#a4a832', marker='+', markersize=20)
# supported values are 'o', '+', '*', '1'
```

```python
# More oprtions for markers
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, marker='o', markersize=20, 
        markerfacecolor='yellow',
        markeredgecolor='red',
       markeredgewidth=5)
```

```python
# Limit plot boundaries (zoom in/out)
ax.set_xlim([-0.5,1])
ax.set_ylim([-0.5,2])
fig
```

### Grid

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

### Twin axes

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

### Plot styles

```python
n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

### Text annotation

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)

ax.text(0.15, 0.2, r"$y=x^2$", fontsize=20, color="blue")
ax.text(0.65, 0.1, r"$y=x^3$", fontsize=20, color="green");
```
