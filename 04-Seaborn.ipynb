# Seaborn

Library for statistical plots.

API: https://seaborn.pydata.org/api.html

### Distribution
* distplot: [DEPRECATED] distribution of an univariate set of observations.
* histplot: Plot univariate or bivariate histograms to show distributions of datasets.
* jointplot: combine two distribution plots.
* pairplot: plot pairwise relationships across the entire dataframe.
* rugplot: draw a dash mark for every point in the distribution. Used to calculate kernel density estimate.

### Categorical
* barplot: aggregate categorical data based on some function (mean by default).
* countplot: count number of occurences.
* boxplot, violinplot: distribution of categorical data.
* stripplot: scatter plot where one variable is categorical.
* swarmplot: combine stripplot and violinplot.
* catplot: uses 'kind' argument to recreate the previous plots.

### Matrix
* headmap: data should already be in a 'matrix' form - column & rows should represent meaninful variables.
* clustermap: matrix dataset as a hierarchically-clustered heatmap.

### Grids
Manually specify grid mapping to create grid charts.

### Regression
Plot regression model charts.

### Style and Color
Borders, font size/scale, colors, context, matplotlib figure parameters.
