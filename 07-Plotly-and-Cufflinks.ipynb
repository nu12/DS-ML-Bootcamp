## Chart types:
* histogram
* scatter
* bar
* boxplot
* 3d surface
* spread
* bubble
* scatter_matrix

```python
import numpy as np
import pandas as pd

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
import chart_studio.plotly as py
```

```python
init_notebook_mode(connected=True) # connect to the javascript of the Notebook
cf.go_offline() # use cufflinks offline
```

```python
# Data 1
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df
```

```python
# Data 2
df2 = pd.DataFrame({'Category': 'A B C'.split(), 'Values': [10, 22, 32]})
df2
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=2)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
# Data 3
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z': [5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
df['A'].iplot(kind='hist', bins=40)
```

```python
df.iplot(kind='hist', bins=40)
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```

```python
df.scatter_matrix()
```
