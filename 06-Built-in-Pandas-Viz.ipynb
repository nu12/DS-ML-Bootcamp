# Pandas built-in visualization
* histogram
* area
* bar
* line
* scatter
* box
* hexbin
* kde/density

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns # changes the look of pandas charts
```

```python
# Time series data
df1 = pd.read_csv('data/06-df1', index_col=0)
df1.head()
```

```python
# Non-time series data
df2 = pd.read_csv('data/06-df2')
df2.head()
```

```python
# Histogram
df1['A'].hist(bins=20)
```

```python
df1['A'].plot(kind='hist')
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.line()
```

```python
df1.plot.line()
```

```python
# Using plt args
df1.plot.line(y='B', figsize=(12,3))
```

```python
# Using plt args
df1.plot.line(y='B', figsize=(12,3), lw=0.5)
```

```python
df1.plot.scatter(x='A', y='B')
```

```python
# c = color
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# s = size
df1.plot.scatter(x='A', y='B', s=df1['C'] * 100, c='C', cmap='coolwarm')
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.plot.hexbin(x='a',y='b')
```

```python
df.plot.hexbin(x='a',y='b', gridsize=25, cmap='magma')
```

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
df2.plot.density()
```
